```python
import math
import numpy as np
import pandas as pd
from itertools import product
import os
from os import listdir
from pathlib import Path
import pickle
import argparse
from typing import List, Tuple
from collections import defaultdict
from MSSDA_main import get_paerser_arguments, run, get_pkl_dic, check_file_str
```

```python
def get_params(args: argparse.Namespace,target_epoch: int,target_measure: int) -> Tuple[str, str, int, float, float]:
    """
    This function parses the args and returns vital information and the result.
    Args:
        args (argparse.Namespace): Command-line arguments for configuration.
        target_epoch (int): the epoch at which the result should be returned.
        target_measure (int): -1 for c-index, and -2 for c-index'.

    Returns:
        Tuple[List[np.ndarray], List[np.ndarray], List[np.ndarray], List[str]]: A tuple containing the following lists:
            - key: List of mRNA features as numpy arrays.
            - domain: List of survival time data as numpy arrays.
            - current_experiment: List of event/censoring status data as numpy arrays.
            - percent_target: List of domain names associated with the data.
            - value: List of domain names associated with the data.
    """
    params = args["args"]
    model, lr, lr_alphas, norm_alphas, target_domain_type = params.model, params.lr, params.lr_alphas, params.norm_alphas, params.target_domain_type
    current_experiment, percent_target = params.current_experiment, params.percent_target
    key = f"{model}_{lr}_{lr_alphas}_{norm_alphas}"
    for _ in new_args.keys():
        if "Before_Learning" in _:
            needed_key = _
            break
    domain = needed_key[:-len("Before_Learning")-1]
    value = args[f"{domain}_{current_experiment}_{target_epoch}"][target_measure]
    return key,domain,current_experiment, percent_target, value
```

## Running on the miRNA dataset

```python
args = get_paerser_arguments()
args.RNA_Type = 0
args.data_path = r"/home/ashaker/ProtoSurvive/data/TCGA/miRNA"
args.save_path = r"/home/ashaker/ProtoSurvive/results/MSSDAmiRNA"
args.lr, args.lr_alphas = 0.0001, 0.001
args.batch_size, args.feat_extract_size, args.nb_experiments = 80, 200, 5
args.largeVal, args.epochs_adapt, args.cuda, args.norm_alphas = 1, 20, True, 2

exps = range(args.nb_experiments)
Test_domain_ids = range(21)
target_domain_types = [1,4]
percent_targets = ["0.00","0.05","0.1","0.15","0.2","0.25"]
percent_targets = ["0.00","0.05"]
model_names = ["MSSDA.SurviveSDI.SurviveSDI"]
```

```python
if os.path.exists(args.save_path):
    already = get_pkl_dic(args.save_path)
else:
    already = {}

Current_index=0
for i, element in enumerate(product(exps,Test_domain_ids,target_domain_types,
                                percent_targets,model_names)):
    current_experiment,Test_domain_id,target_domain_type, percent_target, model = element
    args.current_experiment, args.Test_domain_id  = int(current_experiment), int(Test_domain_id)
    args.target_domain_type = int(target_domain_type)
    args.percent_target = float(percent_target)
    args.model = model
    print("Running The following configuration:",args)
    if args.target_domain_type==1 and float(args.percent_target)!=float("0.00"):
        continue
    if args.target_domain_type!=1 and float(args.percent_target)==float("0.00"):
        continue

    key = str(args.model) +"_"+ str(args.Test_domain_id) +"_"+str(args.target_domain_type) \
        +"_"+str(args.target_domain_weight)+"_"+str(args.percent_target) \
        +"_"+str(args.lr)+"_"+str(args.lr_alphas) \
        +"_"+str(args.norm_alphas)+ "_"+str(args.current_experiment)
    print(key)
    if key not in already.keys() and not check_file_str(args.save_path,key):
        print("Start",key)
        print(args.current_experiment, args.Test_domain_id, args.target_domain_type, args.percent_target, args.model)
        run(args, key)
        print(Current_index)

    Current_index+=1
    print("#############################################################")
```

## Collecting the miRNA results


```python
results_folder = Path(r"/home/ashaker/ProtoSurvive/results/MSSDAmiRNA/Final")
```

```python
pkl_files = [results_folder/f for f in listdir(results_folder) if ".pkl" in f]
parametrization_dic = defaultdict(lambda: defaultdict(lambda : defaultdict(lambda : defaultdict())))
target_measure = -1 #-1 for c-index, and -2 for c-index'.
target_epoch = 19
for pkl_file in pkl_files:
    with open(pkl_file, 'rb') as f:
        unpickler = pickle.Unpickler(f)
        new_args = unpickler.load()
        key,domain,current_experiment, percent_target, value = get_params(new_args,target_epoch,target_measure)
        parametrization_dic[key][domain][percent_target][current_experiment] = value

results = pd.DataFrame()
key = list(parametrization_dic.keys())[0]
for domain in parametrization_dic[key].keys():
    temp_dict = {}
    for percent_target in parametrization_dic[key][domain].keys():

        res = [parametrization_dic[key][domain][percent_target][i] for i in parametrization_dic[key][domain][percent_target].keys()]
        temp_dict[percent_target] =  ["{:.3f}".format(np.mean(res)),"{:.3f}".format(np.std(res)/math.sqrt(5))]

    results  = pd.concat([results, pd.DataFrame(temp_dict, index=[domain,f"{domain}-std"])])
results
```
